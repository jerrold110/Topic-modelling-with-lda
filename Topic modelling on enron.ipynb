```python
import pandas as pd
from nltk import word_tokenize
import numpy as np

np.random.seed(1)
```

```python
cleaned_emails = pd.read_csv('enron_emails_clean.csv')
cleaned_emails.isna().sum()
```

```python
vectors = cleaned_emails.dropna()['cleaned_content'].str.split(" ").values
```

```python
word_data = []
for email in vectors:
    l1 = []
    for word in email:
        l1.append(word)
    word_data.append(l1)
```

```python
from gensim.corpora import Dictionary
```

```python
dictionary = Dictionary(word_data)
print(type(dictionary))

# Filter out frequent and non-frequent word
# Words that appear in at least 20 emails
# Keep and 50000 most frequent words
dictionary.filter_extremes(no_below=20, keep_n=50000)
```

```python
corpus = [dictionary.doc2bow(text) for text in word_data]
corpus[:20]
```

```python
from gensim.models.ldamodel import LdaModel
```

```python
from gensim.models import LdaMulticore
```

```python
ldamodel = LdaMulticore(corpus, num_topics=3, id2word=dictionary, passes=10, workers=4)
```

```python
topics = ldamodel.print_topics(num_words=10)
for topic in topics:
    print(topic)
```

```python
import pyLDAvis.gensim
```

```python
lda_display = pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary, sort_topics=False)

pyLDAvis.display(lda_display)
```
